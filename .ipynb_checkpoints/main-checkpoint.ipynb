## 1. Data import

```python
# import pkgs
%load_ext autoreload
%autoreload 2
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style = 'darkgrid', color_codes = True)
sns.set_context("notebook")
```

```python
# load data
df = pd.read_csv('data/train.csv')
# add column with filenames
df['filename'] = df['id_code'] + '.png' 
df['diagnosis'] = df['diagnosis'].astype('str')

print('Number of labeled images: {}'.format(df.shape[0]))
print('Number of classis: {}'.format(len(set(df['diagnosis']))))
df.head()
```

## 2. Data Exploration

```python
# Plot example images covering all 5 levels of diagnosis
def plot_examples(idx = 0): 
    fig, axs = plt.subplots(1, 5, figsize = (16,16))
    for ctr, label in enumerate(np.sort(df['diagnosis'].unique())):
        # filter label
        file = df[df['diagnosis'] == label].iloc[idx,2]
        img = cv2.imread('data/train_images/' + file)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        h, w,_ = img.shape
        # h, w = img.shape # for GRAY
        axs[ctr-1].imshow(img, cmap = 'gray')
        axs[ctr-1].set_title('Diagnosis {}\n(h{} x w{})'.format(label,h,w), fontweight = 'bold')
        axs[ctr-1].set_xticks([])
        axs[ctr-1].set_yticks([])
    plt.tight_layout()
    plt.show()
    
# call fun
f = np.vectorize(plot_examples)
f(np.random.randint(0, 200, size = 4))
```

```python
# Figure relative distribution
fig = plt.figure(figsize = (10,6))
ax = sns.barplot(x = "diagnosis",  y = "diagnosis", data = df, estimator = lambda x: len(x) / len(df))
# add numbers on top
for p in ax.patches:
    ax.text(x = p.get_x() + p.get_width() / 2,
            y = p.get_height() * 1.01,
            s = 'N = {:1.0f}'.format(p.get_height() * df.shape[0]),
            ha = "center", fontsize = 12) 
ax.set_xticklabels(['No DR', 'Mild', 'Moderate', 'Severe', 'Proliferative DR'], rotation = 0)
plt.ylabel("Relative frequency")
plt.xlabel("Diagnosis")
plt.title('Data distribution', fontweight = 'bold')
plt.show()
```

## 3. Data preprocessing

```python
from sklearn.model_selection import train_test_split
# Split data into train, test, and validation
train, test = train_test_split(df, test_size = 0.30, random_state = 42, shuffle = True)
train, vali  = train_test_split(train, test_size = 0.25, random_state = 42, shuffle = True)

print('Images:\n Train: {}\n Test: {} \n Validation: {}'.format(train.shape[0], test.shape[0], vali.shape[0]))
```

```python
from keras.preprocessing.image import ImageDataGenerator
# Constrcut data generator for image flow and data augmentation
# (from https://keras.io/preprocessing/image/)
# TODO show examples
datagen = ImageDataGenerator(rescale = 1./255.,
                             zoom_range = 0.2,                   
                             width_shift_range = 0.1,
                             height_shift_range = 0.1,
                             horizontal_flip = True)

df_classes = ['0', '1', '2', '3', '4']
img_rescale = (224, 224)

# train
train_gen = datagen.flow_from_dataframe(dataframe = train,
                                              x_col = 'filename',
                                              y_col = 'diagnosis',
                                              directory = 'data/train_images/',
                                              target_size = img_rescale,
                                              batch_size = 32,
                                              color_mode = 'rgb',
                                              classes = df_classes)
# val
validation_gen = datagen.flow_from_dataframe(dataframe = vali,
                                              x_col = 'filename',
                                              y_col = 'diagnosis',
                                              directory = 'data/train_images/',
                                              target_size = img_rescale,
                                              batch_size = 32,
                                              color_mode = 'rgb',
                                              classes = df_classes)

# Data generator without augmentation
test_datagen = ImageDataGenerator(rescale = 1./255.)
test_gen = test_datagen.flow_from_dataframe(dataframe = train,
                                              x_col = 'filename',
                                              y_col = 'diagnosis',
                                              directory = 'data/train_images/',
                                              target_size = img_rescale,
                                              batch_size = 32,
                                              color_mode = 'rgb',
                                              classes = df_classes)
```

### 4. Naive Model

```python
from sklearn.metrics import precision_recall_fscore_support

# naive model label every image as class 0
naive_pred = np.repeat('0', test.shape[0])

# compute 
precision, recall, _, _ = precision_recall_fscore_support(test['diagnosis'], naive_pred, average = 'micro')

print('Precision: {:1.2f}'.format(precision))
print('Recall:    {:1.2f}'.format(recall))
```

### 5. ResNet50 Model

```python
from keras.applications.resnet50 import ResNet50
from keras.models import Model
from keras.layers import Input, Conv2D, MaxPooling2D, GlobalAveragePooling2D, Dropout, Flatten, Dense, BatchNormalization

def get_resnet_model(img_shape, num_class): 
    input_img = Input(shape = img_shape) 
    resnet50 = ResNet50(weights = 'imagenet', include_top = False, input_tensor = input_img)
    x = GlobalAveragePooling2D()(resnet50.output)
    x = BatchNormalization()(x)
    x = Dropout(0.2)(x)
    x = Dense(1024)(x)
    x = BatchNormalization()(x)
    x = Dropout(0.2)(x)
    x = Dense(512)(x)
    output = Dense(num_class, activation= 'softmax')(x)
    return Model(input_img, output)

model_resnet = get_resnet_model((224,224,3),5)
model_resnet.summary()

# set trainable layers
for layers in model_resnet.layers:
    layers.trainable =  True

# compile model
model_resnet.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint, LearningRateScheduler

# Checkpointer
checkpointer = ModelCheckpoint(filepath = 'models/weights.best.Resnet50.hdf5', 
                               save_weights_only =  True,
                               save_best_only = True,
                               verbose = 1)
```

```python
# train model
model_resnet.fit_generator(
        train_gen,
        steps_per_epoch = 2000 // 32,
        epochs = 2,
        validation_data = validation_gen,
        validation_steps = 600 // 32,
        callbacks = [checkpointer])
```
