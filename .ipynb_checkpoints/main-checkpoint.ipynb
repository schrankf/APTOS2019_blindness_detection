### 1. Data import

```python
# import pkgs
%load_ext autoreload
%autoreload 2
import pandas as pd
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from sklearn.model_selection import train_test_split
from keras.preprocessing.image import ImageDataGenerator

sns.set(style = 'ticks', color_codes = True)
%matplotlib inline
```

```python
# get data
df = pd.read_csv('data/train.csv')
# add column with filenames
df['filename'] = df['id_code'] + '.png' 
df['diagnosis'] = df['diagnosis'].astype('str')

print('Number of labeled images are {}'.format(df.shape[0]))
print(df.dtypes)
```

### 2. Data Exploration

```python
# Plot example images covering all levels of diagnosis
fig, axs = plt.subplots(1, 5, figsize = (16,16))

for ctr, label in enumerate(np.sort(df['diagnosis'].unique())):
    # filter label
    file = df[df['diagnosis'] == label].iloc[0,2]
    img = cv2.imread('data/train_images/' + file)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    h, w = img.shape
    axs[ctr-1].imshow(img, cmap = 'gray')
    axs[ctr-1].set_title('Diagnosis {}\n(h{} x w{})'.format(label,h,w), fontweight = 'bold')
    axs[ctr-1].set_xticks([])
    axs[ctr-1].set_yticks([])
    
plt.tight_layout()
plt.show()
```

```python
# Figure relative distribution
fig = plt.figure(figsize = (8,5))
sns.barplot(x = "diagnosis",  y = "diagnosis", data = df, estimator = lambda x: len(x) / len(df))
plt.ylabel("Relative frequency")
plt.title('Data distribution', fontweight = 'bold')
plt.show()
```

### 3. Data preprocessing

```python
# Split data into train, test, and validation
train, test = train_test_split(df, test_size = 0.25, random_state = 42, shuffle = True)
train, vali  = train_test_split(train, test_size = 0.25, random_state = 42, shuffle = True)

print('Images:\n Train: {}\n Test: {} \n Validation: {}'.format(train.shape[0], test.shape[0], vali.shape[0]))
```

```python
# Constrcut data generator for image flow and data augmentation
# (from https://keras.io/preprocessing/image/)
datagen = ImageDataGenerator(rescale = 1./255.,
                             zoom_range = 0.2,                   
                             width_shift_range = 0.1,
                             height_shift_range = 0.1,
                             horizontal_flip = True)
df_classes = ['0', '1', '2', '3', '4']
# train
train_gen = datagen.flow_from_dataframe(dataframe = train,
                                              x_col = 'filename',
                                              y_col = 'diagnosis',
                                              directory = 'data/train_images/',
                                              target_size = (244, 244),
                                              batch_size = 32,
                                              color_mode = 'grayscale',
                                              classes = df_classes)
# val
validation_gen = datagen.flow_from_dataframe(dataframe = validation,
                                              x_col = 'filename',
                                              y_col = 'diagnosis',
                                              directory = 'data/train_images/',
                                              target_size = (244, 244),
                                              batch_size = 32,
                                              color_mode = 'grayscale',
                                              classes = df_classes)

# Data generator without augmentation
test_datagen = ImageDataGenerator(rescale = 1./255.)
test_gen = test_datagen.flow_from_dataframe(dataframe = train,
                                              x_col = 'filename',
                                              y_col = 'diagnosis',
                                              directory = 'data/train_images/',
                                              target_size = (244, 244),
                                              batch_size = 32,
                                              color_mode = 'grayscale',
                                              classes = df_classes)
```
